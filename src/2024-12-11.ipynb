```python
from datetime import datetime
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

- トヨタ: 7203
- ファーストリテイリング: 9983
- 日立製作所: 6501
- KDDI: 9433

```python
tickers = ['7203.T', '9983.T', '6501.T', '9433.T']
start = datetime(2010, 1, 1)
end = datetime(2024, 12, 11)

df = yf.download(tickers, start=start, end=end)
```

## 日次リターン

```python
fig, axs = plt.subplots(len(tickers), 1, figsize=(12, 6 * len(tickers)))

for i, ticker in enumerate(tickers):
    axs[i].plot(np.log(df['Adj Close'][ticker]).diff(), label=ticker)
    axs[i].legend()

plt.tight_layout()
plt.show()
```
